```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
movies_df = pd.read_csv('../netflix_titles.csv')
```

```python
movies_df.sort_values(by='release_year', ascending=False)
```

```python
df_2021 = movies_df[movies_df['release_year'] == 2021]
```

```python
anos_interesse = [2018, 2019, 2020, 2021]
paises_interesse = ['Brazil', 'United States', 'India', 'Italy', 'United Kingdom']
df_filtrado = movies_df[(movies_df['country'].isin(paises_interesse)) & (movies_df['release_year'].isin(anos_interesse))]

# Contar a quantidade de itens para cada combinação de 'release_year' e 'country'
contagem = df_filtrado.groupby(['release_year', 'country']).size().unstack().fillna(0)

# Ordenar as colunas em ordem crescente pelo número de itens
contagem = contagem[contagem.sum().sort_values(ascending=True).index]

# Criar um gráfico de barras
ax = contagem.plot(kind='bar', stacked=False)

# Adicionar rótulos e título ao gráfico
plt.xlabel('Ano de Lançamento')
plt.ylabel('Quantidade de Itens')
plt.title('Gráfico de Barras de Filmes e Séries da Netflix (Ordem Crescente)')

# Adicionar valores nas barras
for i in ax.patches:
    ax.annotate(str(i.get_height()), (i.get_x() + i.get_width() / 2, i.get_height()), ha='center', va='bottom')

# Mostrar o gráfico
plt.show()
```
